# Prueba de Animaciones con Matplotlib

Este notebook sirve para probar animaciones reales con `matplotlib.animation.FuncAnimation` en VS Code.

Objetivo:
- Verificar si VS Code reproduce `to_jshtml()`
- Comparar una animación simple con una animación de grafo
- Evitar el enfoque de múltiples `display(fig)` que solo apila imágenes


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, display
import networkx as nx

plt.rcParams['figure.dpi'] = 110
plt.rcParams['animation.embed_limit'] = 50

print('Si VS Code soporta jshtml, abajo deberías ver una animación real y no imágenes repetidas.')
```

## 1. Animación simple

Esta prueba es mínima: una onda seno y un punto que se desplaza.
Si esta no se anima, el problema es el frontend de VS Code y no tu notebook principal.

```python
x = np.linspace(0, 2 * np.pi, 300)\ny = np.sin(x)\n\nfig, ax = plt.subplots(figsize=(7, 3.5))\nfig.patch.set_facecolor('white')\nax.set_facecolor('#F8FBFF')\nax.plot(x, y, color='#2563EB', linewidth=2.5)\ndot, = ax.plot([], [], 'o', color='#DC2626', markersize=10)\ntext = ax.text(0.02, 0.92, '', transform=ax.transAxes, fontsize=11, color='#111827')\nax.set_xlim(x.min(), x.max())\nax.set_ylim(-1.2, 1.2)\nax.set_title('Prueba 1: Onda seno animada', fontsize=13)\nax.grid(alpha=0.25, linestyle='--')\n\ndef init_sine():\n    dot.set_data([], [])\n    text.set_text('frame=0')\n    return dot, text\n\ndef update_sine(frame):\n    idx = frame % len(x)\n    dot.set_data([x[idx]], [y[idx]])\n    text.set_text(f'frame={frame}')\n    return dot, text\n\nani_sine = FuncAnimation(\n    fig, update_sine, init_func=init_sine,\n    frames=120, interval=40, blit=True, repeat=True\n)\nplt.close(fig)\ndisplay(HTML(ani_sine.to_jshtml()))\n
```

## 2. Animación de grafo pequeña

Aquí simulamos una idea parecida a la de tu notebook: un nodo central resaltado y vecinos con pulsación suave.

```python
G = nx.Graph()\nG.add_edges_from([\n    ('nuevo', 'v1'), ('nuevo', 'v2'), ('nuevo', 'v3'),\n    ('v1', 'v4'), ('v2', 'v5'), ('v3', 'v6'),\n    ('v4', 'v5'), ('v5', 'v6')\n])\n\npos = nx.spring_layout(G, seed=42, k=0.9, iterations=100)\nneighbors = {'v1', 'v2', 'v3'}\n\nfig, ax = plt.subplots(figsize=(6.5, 5.5))\nfig.patch.set_facecolor('white')\n\ndef draw_graph_frame(frame):\n    ax.clear()\n    ax.set_facecolor('#F8FBFF')\n    pulse = 1.0 + 0.18 * np.sin(frame / 4)\n\n    nx.draw_networkx_edges(\n        G, pos, ax=ax, edge_color='#94A3B8', width=1.5, alpha=0.45\n    )\n\n    base_nodes = [n for n in G.nodes if n not in neighbors and n != 'nuevo']\n    nx.draw_networkx_nodes(\n        G, pos, nodelist=base_nodes, ax=ax, node_color='#CBD5E1',\n        node_size=420, edgecolors='white', linewidths=1.2\n    )\n\n    nx.draw_networkx_nodes(\n        G, pos, nodelist=list(neighbors), ax=ax, node_color='#60A5FA',\n        node_size=520 * pulse, edgecolors='#1D4ED8', linewidths=2.0\n    )\n\n    nx.draw_networkx_nodes(\n        G, pos, nodelist=['nuevo'], ax=ax, node_color='#F97316',\n        node_size=820 * pulse, node_shape='*', edgecolors='#9A3412', linewidths=2.0\n    )\n\n    nx.draw_networkx_labels(G, pos, ax=ax, font_size=10, font_weight='bold')\n    ax.set_title('Prueba 2: Subgrafo local animado', fontsize=13)\n    ax.text(0.02, 0.02, 'Estrella = texto nuevo | Azul = vecinos ancla',\n            transform=ax.transAxes, fontsize=9, color='#334155',\n            bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='#CBD5E1'))\n    ax.axis('off')\n\nani_graph = FuncAnimation(fig, draw_graph_frame, frames=90, interval=70, repeat=True)\nplt.close(fig)\ndisplay(HTML(ani_graph.to_jshtml()))\n
```

## Cómo interpretar el resultado

- Si ves una animación real aquí, entonces podemos llevar `to_jshtml()` al notebook principal.
- Si aquí también ves imágenes repetidas o nada, entonces VS Code no está reproduciendo bien este formato y conviene usar una visualización estática más fuerte.
