## 1. Introduccion: El Ecosistema de la Inteligencia Artificial y la Evolucion hacia las GNN

Para comprender el posicionamiento metodologico de **SentinelGrafo**, es fundamental mapear conceptualmente el ecosistema de la Inteligencia Artificial y entender en que punto las arquitecturas tradicionales encuentran sus limites teoricos al enfrentarse a datos interconectados.

A continuacion, se presenta la jerarquia taxonomica que rige el desarrollo del aprendizaje profundo moderno, ubicando de forma precisa el rol de las Redes Neuronales de Grafos (GNN):

![Taxonomia IA hacia GNN](../06_Recursos/taxonomia_ia_gnn.png)

### 1.1 Desglose del Ecosistema y Campos de Aplicacion

* **Inteligencia Artificial (IA):** El marco macro que engloba cualquier sistema computacional disenado para simular capacidades cognitivas humanas.
* **Machine Learning (ML):** Subcampo enfocado en algoritmos estadisticos que permiten a las maquinas aprender patrones directamente a partir de los datos sin ser programadas explicitamente.
* **Deep Learning (DL):** Evolucion del ML basada en el uso de arquitecturas jerarquicas multicapa (redes neuronales artificiales) capaces de extraer abstracciones de alto nivel de forma automatizada.

Dentro del espectro de las **Redes Neuronales Artificiales**, la especializacion estructural dicta su exito segun el dominio del problema:

* **Perceptron Multicapa (MLP) / Redes Simples:** Disenadas para procesar vectores estaticos independientes. Su fuerte radica en los **datos tabulares** convencionales (registros financieros, analitica de clientes o predicciones estructuradas).
* **Redes Neuronales Convolucionales (CNN):** Especializadas en datos con topologia euclidiana bidimensional regular, aprovechando la invarianza espacial para dominar la **vision por computadora** (procesamiento de imagenes y video).
* **Redes Neuronales Recurrentes (RNN, LSTM, GRU):** Estructuradas para capturar dependencias temporales y secuenciales en flujos unidireccionales de datos, como **series de tiempo y senales**.
* **Transformers:** Basados en mecanismos de auto-atencion masiva, revolucionaron el procesamiento de secuencias complejas, convirtiendose en el motor detras del **procesamiento de lenguaje moderno (LLMs como ChatGPT)** y traduccion de documentos.
* **Redes Neuronales de Grafos (GNN):** El paradigma de vanguardia disenado especificamente para operar sobre datos relacionales estructurados en dominios no euclidianos. Son la herramienta idonea para modelar **redes sociales, sistemas de recomendacion, rutas logisticas y conexiones semanticas**.

### 1.2 Limitaciones del Aprendizaje Clasico ante Datos Relacionales

El aprendizaje automatico clasico ha cosechado exitos notables en el procesamiento de datos estructurados en dominios euclidianos bidimensionales o secuenciales, tales como imagenes o series temporales. Sin embargo, su suposicion matematica fundamental —que las observaciones son independientes e identicamente distribuidas ($i.i.d.$)— colapsa ante la presencia de datos relacionales complejos representados como grafos.

En el analisis de redes, las entidades no existen de forma aislada; por el contrario, sus caracteristicas y comportamientos estan intrinsecamente correlacionados a traves de enlaces topologicos. Al aplicar un Perceptron Multicapa (MLP) tradicional a tareas de clasificacion de nodos, se comete el error metodologico de asumir la independencia de las observaciones. Un MLP procesa cada vector de atributos de nodo de forma aislada, lo que imposibilita la integracion directa de las caracteristicas del vecindario estructurado o de la conectividad de la red en la prediccion.

Intentar alimentar un MLP con la estructura de un grafo aplanando la matriz de adyacencia en un vector de entrada rigido introduce severas limitaciones teoricas:

* La incapacidad de operar con grafos de tamano variable.
* La perdida de la propiedad de invariancia ante permutaciones de los nodos.
* Una explosion de parametros inmanejable para el optimizador.

Para cerrar la brecha entre estas representaciones discretas y el aprendizaje continuo, **SentinelGrafo** adopta el paradigma de *graph embedding*, proyectando nodos y textos en un espacio vectorial continuo de baja dimensionalidad $\mathbb{R}^d$. En este espacio latente, la geometria euclidiana preserva las similitudes semanticas, las propiedades de conectividad estructural y los atributos de la red original, actuando como un puente optimo para superar las limitaciones lineales y alcanzar clasificaciones de alta precision.

## 2. Resumen

Este trabajo presenta el desarrollo e implementacion de SentinelGrafo, un software especializado en la clasificacion de mensajes criticos y noticias en plataformas digitales mediante el uso de Redes Neuronales Graficas (GNN). A diferencia del aprendizaje automatico clasico que asume independencia en los datos, SentinelGrafo transforma texto no estructurado en estructuras relacionales complejas (grafos dispersos) mediante vectorizacion TF-IDF y conectividad por vecindad adaptativa ($k$-NN).

Se evaluan de forma comparativa las arquitecturas GraphSAGE (enfoque inductivo basado en muestreo local) y GCN (convolucion espectral aproximada) sobre dos entornos criticos: la deteccion de tweets de desastres reales (Disaster Tweets) y la categorizacion masiva de noticias internacionales (AG News), demostrando como el paradigma de Message Passing captura dependencias semanticas y topologicas superiores a los modelos lineales planos (Baselines).

## 4. Marco teorico

### 4.1. Redes Neuronales Graficas (GNN) y el mecanismo de Message Passing

Las Redes Neuronales Graficas (GNN) resuelven las deficiencias de los clasificadores planos mediante la implementacion del paradigma de Paso de Mensajes (Message Passing). El principio operativo de este paradigma dicta que la representacion latente de cada nodo individual se actualiza de manera iterativa al recopilar, transformar y consolidar informacion estructurada proveniente de su entorno local.

Formalmente, el paso de mensajes en una capa de GNN $k$ se rige por un proceso de dos etapas estructurado matematicamente para cada nodo $v \in \mathcal{V}$:

1. **Etapa de Agregacion (AGGREGATE):** El nodo central recolecta las representaciones de sus vecinos inmediatos $\mathcal{N}(v)$ provenientes de la iteracion previa $(k-1)$. La funcion de agregacion $f$ debe ser un mapeo invariante ante permutaciones de los vectores del conjunto (multiconjunto), asegurando que el orden arbitrario de indexacion de los vecinos no afecte el resultado final:
   $$m_v^{(k)} = AGGREGATE^{(k)} (\{h_u^{(k-1)} : u \in \mathcal{N}(v)\})$$

   donde $h_u^{(k-1)}$ representa el vector de caracteristicas latentes del nodo vecino $u$ en la capa $(k-1)$.

2. **Etapa de Actualizacion / Combinacion (COMBINE):** El mensaje acumulado $m_v^{(k)}$ se fusiona con la representacion del propio nodo central de la capa anterior $h_v^{(k-1)}$ mediante una funcion diferenciable $\phi$ (como un perceptron con funciones de activacion no lineales) para proyectar el nuevo estado latente $h_v^{(k)}$:
   $$h_v^{(k)} = COMBINE^{(k)} (h_v^{(k-1)}, m_v^{(k)})$$

Al apilar recursivamente K capas de paso de mensajes, el modelo expande progresivamente el campo receptivo de cada nodo de manera analoga a como las capas de una Red Neuronal Convolucional (CNN) tradicional incrementan el tamano de sus kernels. En la capa K, el embedding de salida $h_v^{(K)}$ codifica no solo las propiedades semanticas intrinsecas del nodo v, sino tambien la topologia relacional y la distribucion de caracteristicas de todo su vecindario a una distancia de K saltos topologicos (hops).

### 4.2. Graph Convolutional Networks (GCN)

La arquitectura Graph Convolutional Network (GCN) generaliza las operaciones de convolucion espacial de imagenes a estructuras de grafos arbitrarias mediante una simplificacion de primer orden de las convoluciones espectrales basadas en la teoria de senales sobre grafos.

En la teoria espectral, las senales de un grafo se filtran en el dominio de la frecuencia mediante la descomposicion en componentes de Fourier del Laplaciano normalizado del grafo. Dado que calcular los autovectores del Laplaciano para realizar la transformada de Fourier es extremadamente costoso computacionalmente, Kipf y Welling introdujeron una aproximacion localizada utilizando polinomios de Chebyshev truncados de primer orden, asumiendo una escala espectral constante.

Bajo estas aproximaciones de filtrado espacial localizado, la operacion de una capa GCN convolucional estandar se puede formular de forma compacta como:
$$H^{(l+1)} = \sigma(\hat{A} H^{(l)} W^{(l)})$$

donde $W^{(l)}$ es la matriz de pesos entrenable para la capa $l$, y $\sigma$ es una funcion de activacion no lineal (tipicamente ReLU). Sin embargo, la matriz de adyacencia pura $A$ presenta un problema grave: carece de bucles propios en la diagonal ($A_{ii} = 0$), lo que significa que el estado actualizado de un nodo $v_i$ se calcula sumando exclusivamente la informacion de sus vecinos externos, ignorando por completo su propia senal original.

Para mitigar estos problemas, Kipf y Welling formularon el **Truco de Renormalizacion**. Este metodo anade bucles propios explicitos a la estructura mediante la definicion de la matriz de adyacencia aumentada $\tilde{A} = A + I$, calculando su correspondiente matriz diagonal de grados aumentada $\tilde{D}$. Con este ajuste estructural, la ecuacion de propagacion final se define como:
$$H^{(l+1)} = \sigma(\tilde{D}^{-1/2} \tilde{A} \tilde{D}^{-1/2} H^{(l)} W^{(l)})$$

La trascendencia teorica de este truco radica en que la insercion de bucles propios reduce el espectro del Laplaciano normalizado, contrayendo su autovalor maximo de $\lambda_{max} \approx 2$ a aproximadamente $\tilde{\lambda}_{max} < 1.5$. Esta contraccion elimina los efectos desestabilizadores de los coeficientes de alta frecuencia de la senal. En consecuencia, la matriz de propagacion renormalizada actua como un **filtro de paso bajo (low-pass filter)** que suaviza de manera sumamente efectiva las caracteristicas vecinas a lo largo de las aristas del grafo. Es la opcion ideal para grafos estaticos y fuertemente homofilos, donde el contexto global de conectividad permanece invariable durante el entrenamiento. Su mayor vulnerabilidad teorica es el sobresuavizado (over-smoothing): si se apilan mas de 2 o 3 capas de convolucion isotropica, las representaciones de los nodos convergen a un valor promedio uniforme, perdiendo toda capacidad de diferenciacion predictiva.

### 4.3. GraphSAGE (Sample and Aggregate)

A diferencia de los modelos espectrales tradicionales que asumen un grafo estatico de escala constante (aprendizaje transductivo), GraphSAGE (Sample and Aggregate) propone un marco disenado especificamente para escenarios inductivos. Esto permite generar embeddings representativos para nodos completamente nuevos o subgrafos dinamicos no observados durante la fase de entrenamiento sin necesidad de re-optimizar la red.

La innovacion operativa clave que permite a GraphSAGE escalar eficientemente a grafos de escala masiva (como redes sociales o sistemas de recomendacion con miles de millones de nodos) es el **Muestreo de Vecindarios (Neighborhood Sampling)**. En lugar de procesar la totalidad de las conexiones de un nodo central (lo que generaria un crecimiento exponencial intratable del tamano del subgrafo de computacion con cada capa adicional), GraphSAGE restringe la propagacion a un conjunto muestreado de tamano uniforme y fijo. Matematicamente, para cada nodo central $v$, el vecindario muestreado $\mathcal{N}_{sample}(v)$ se extrae mediante una seleccion aleatoria uniforme de tamano constante $S$ en cada capa $l$:
$$\mathcal{N}_{sample}^{(l)}(v) \sim \text{Uniforme}(\mathcal{N}(v), S)$$

Esto acota la complejidad temporal y espacial por lote de manera rigurosa a $O(B \cdot S^K)$, donde $B$ es el tamano del lote de nodos centrales seleccionados, permitiendo el entrenamiento distribuido e incremental mediante mini-lotes convencionales.

Una vez recolectados los subconjuntos uniformes de vecinos, GraphSAGE ejecuta la agregacion y actualizacion utilizando tres tipos de funciones de agregacion simetricas diferenciables:

- **Agregador de Media (Mean):** Calcula la media aritmetica de los estados de los vecinos directos muestreados, actuando como una aproximacion local lineal simple pero altamente estable.

- **Agregador de Agrupacion (Pooling):** Aplica una transformacion lineal independiente a cada vector de la vecindad muestreada, seguida de una operacion no lineal y un operador simetrico de maximizacion (max-pooling) por coordenadas para aislar las caracteristicas criticas de la vecindad.

- **Agregador LSTM:** Ofrece una capacidad expresiva significativamente mayor para capturar patrones relacionales no lineales complejos. Para contrarrestar la sensibilidad secuencial de las arquitecturas recurrentes de memoria, GraphSAGE aplica de forma simetrica una permutacion aleatoria $\pi$ a los elementos del vecindario muestreado antes de alimentarlos a la celda LSTM, eliminando cualquier sesgo posicional artificial.

Finalmente, el paso de actualizacion de caracteristicas de GraphSAGE se distingue de GCN al evitar la suma directa y optar por la concatenacion explicita del estado previo del nodo central con el vector consolidado del vecindario agrupado, previniendo la degradacion rapida de la identidad semantica del nodo central y atenuando el sobresuavizado:
$$h_v^{(k)} = \sigma(W^{(k)} \cdot [h_v^{(k-1)} \| m_{\mathcal{N}(v)}^{(k)}])$$

### 4.4. Graph Attention Networks (GAT y GATv2)

Las Graph Attention Networks (GAT) introducen un mecanismo de paso de mensajes puramente anisotropico, superando la limitacion de los filtros isotropicos (como GCN o promedios simples) donde las senales se promedian de manera uniforme e inflexible basandose exclusivamente en el grado topologico del nodo. Al emplear auto-atencion (self-attention) localizada y enmascarada, GAT permite que cada nodo calcule dinamicamente un peso de atencion adaptativo sobre cada uno de sus vecinos, priorizando las aristas semanticamente informativas y aislando las conexiones ruidosas.

En una capa GAT clasica, para un nodo central $i$ y un vecino $j$, se calcula en primer lugar un coeficiente de atencion no normalizado $e_{ij}$ mediante una unica capa lineal parametrizada por un vector de proyeccion aprendible $\mathbf{a}$ sobre la concatenacion de sus senales proyectadas linealmente mediante la matriz de pesos compartida $\mathbf{W}$:
$$e_{ij} = \text{LeakyReLU}(\mathbf{a}^T [\mathbf{W} h_i \| \mathbf{W} h_j])$$

Posteriormente, estos coeficientes se normalizan localmente utilizando una funcion softmax con el fin de asegurar que los pesos de ponderacion sean comparables e invariantes frente a variaciones extremas en los grados de los nodos:
$$\alpha_{ij} = \frac{\exp(e_{ij})}{\sum_{k \in \mathcal{N}(i)} \exp(e_{ik})}$$

El estado final del nodo central se actualiza mediante una suma ponderada no lineal por dichos factores de atencion normalizados:
$$h_i' = \sigma\left(\sum_{j \in \mathcal{N}(i)} \alpha_{ij} \mathbf{W} h_j\right)$$

Para estabilizar el proceso de auto-atencion y robustecer la convergencia del entrenamiento, GAT apila de forma paralela $K$ cabezas de atencion independientes. Para capas intermedias u ocultas, los resultados se concatenan, generando una representacion rica de dimensionalidad $K \cdot d'$:
$$h_i' = \|_{k=1}^K \sigma\left(\sum_{j \in \mathcal{N}(i)} \alpha_{ij}^{(k)} \mathbf{W}^{(k)} h_j\right)$$

Mientras que para la capa de salida o de clasificacion, se aplica un promedio aritmetico simple para evitar el sobredimensionamiento parametrico del modelo.

No obstante, un analisis teorico riguroso desarrollado por Brody, Alon e Yahav revelo que la ecuacion del GAT clasico presenta un defecto estructural de diseno critico. Al aplicar el vector de proyeccion lineal $\mathbf{a}$ directamente despues de concatenar los vectores proyectados individualmente por $\mathbf{W}$, la ecuacion de los logits se descompone de forma estrictamente aditiva dentro del argumento:
$$e_{ij} = \text{LeakyReLU}(\mathbf{a}_L^T \mathbf{W} h_i + \mathbf{a}_R^T \mathbf{W} h_j)$$

donde $\mathbf{a}_L$ y $\mathbf{a}_R$ representan las mitades izquierda y derecha de $\mathbf{a}$ respectivamente. Dado que la funcion de activacion LeakyReLU es monotona, cualquier comparacion de coeficientes entre vecinos preserva estrictamente el orden inducido exclusivamente por el termino dependiente del vecino $\mathbf{a}_R^T \mathbf{W} h_j$, anulando cualquier influencia real de las caracteristicas del nodo central de consulta $i$ en la ordenacion de importancia de sus vecinos. Esto significa que la atencion en el GAT tradicional de Velickovic et al. es de caracter **estatico**: existe un orden de vecinos global determinado exclusivamente por los nodos emisores (claves), el cual es incapaz de reconfigurarse dinamicamente en funcion de las caracteristicas del receptor (consulta).

Para subsanar esta debilidad estructural, se propone **GATv2**, una version optimizada que reordena las operaciones algebraicas internas aplicando la proyeccion lineal de atencion *despues* de haber sometido la representacion de la arista a la transformacion conjunta y su no linealidad asociada:
$$e_{ij} = \mathbf{a}^T \text{LeakyReLU}(\mathbf{W} [h_i \| h_j])$$

En GATv2, al forzar a la matriz lineal unificada $\mathbf{a}$ a actuar directamente sobre la concatenacion integrada de ambos nodos previo a la activacion, la no linealidad LeakyReLU actua mezclando estrechamente las caracteristicas cruzadas de ambos nodos. Esto convierte a GATv2 en un aproximador universal sobre el espacio de funciones de atencion relacionales, desbloqueando una atencion verdaderamente **dinamica** donde la relevancia y jerarquia de los vecinos se reestructura de forma sumamente precisa segun el nodo central de consulta.

## 5. Comparativa tecnica de arquitecturas GNN

Para facilitar la evaluacion heuristica y el diseno de los experimentos de clasificacion del proyecto, se consolida a continuacion una comparativa detallada y sistematica de los modelos analizados.

| Modelo | Tipo de Convolucion / Agregacion | Naturaleza de la Inferencia | Fortaleza Principal | Complejidad Temporal (por capa l) | Tratamiento de Vecindad |
| :---- | :---- | :---- | :---- | :---- | :---- |
| **GCN** | Espacial con aproximacion espectral de primer orden | Principalmente Transductiva (requiere Laplaciano estatico del grafo completo) | Simplicidad parametrica; actua como un filtro low-pass altamente eficaz en dominios homofilos | $O(|\mathcal{E}|)$ | Completa estructurada: todas las aristas en cada capa |
| **GraphSAGE** | Agregacion local no lineal de vecindad muestreada | Inductiva pura (generaliza de forma nativa a nodos y grafos dinamicos no observados) | Excepcional escalabilidad mediante muestreo acotado; idoneo para entrenamiento por mini-lotes | $O(B \cdot S^K)$ | **Muestreado Estructurado:** Subconjunto aleatorio de tamano uniforme fijo ($S$) |
| **GAT** | Auto-atencion relacional localizada y enmascarada | Transductiva e Inductiva | Ponderacion adaptativa de aristas; gran inmunidad frente a conexiones estructurales ruidosas | $O(|\mathcal{V}| \cdot F' + |\mathcal{E}| \cdot F')$ | Completa ponderada: todas las aristas con pesos de atencion |
| **GATv2** | Auto-atencion relacional con mezcla cruzada de caracteristicas | Transductiva e Inductiva | Maxima expresividad matematica; aproximador de atencion dinamico universal | $O(|\mathcal{V}| \cdot F' + |\mathcal{E}| \cdot F')$ | Completa ponderada: atencion sobre la mezcla cruzada de consulta y clave |

La comprension profunda de estas complejidades revela dependencias estructurales criticas: mientras que GCN computa de forma lineal y veloz sobre el conjunto global de aristas $|\mathcal{E}|$, sufre inestabilidades severas de sobresuavizado en estructuras complejas. GAT y GATv2 introducen una sobrecarga computacional asociada a la multiplicacion de matrices y la normalizacion local de softmax sobre las conexiones del grafo. Por ende, en grafos homogeneos densos donde la variabilidad relacional es nula, GCN exhibe tiempos de convergencia sustancialmente inferiores, reservando el poder expresivo anisotropico de GATv2 para dominios relacionales heterogeneos o ruidosos. Por ultimo, GraphSAGE es el unico modelo capaz de asegurar la viabilidad de entrenamiento sobre CPU para grafos a gran escala gracias a la desacoplacion parametrica que impone el factor de muestreo $S$.

## 6. Implicancias metodologicas para los casos de estudio

Las caracteristicas teoricas delineadas para cada arquitectura establecen una clara hoja de ruta tecnica para los dos experimentos computacionales planteados en el plan de trabajo del Grupo Segovia.

### Caso 1: Deteccion de tweets sobre desastres reales (Disaster Tweets)

El conjunto de datos Disaster Tweets describe una coleccion de mensajes con una topologia construida artificialmente mediante similitud coseno TF-IDF y alta homofilia semantica.

- **Comportamiento esperado de GCN:** Actuara como un suavizador de paso bajo idoneo para este escenario. Al aplicar el truco de renormalizacion, el modelo filtrara eficientemente las caracteristicas de frecuencia de palabras, logrando una demarcacion clara de las categorias en las capas ocultas y sirviendo como una opcion fuerte y eficiente. Su debilidad ocurrira en mensajes ambiguos o con terminologia atipica.
- **Comportamiento esperado de GAT / GATv2:** Permitira al clasificador asignar pesos atenuados a conexiones semanticamente marginales o ruidosas, concentrando la auto-atencion en los vecinos mas informativos. Al implementar GATv2, el modelo sera capaz de capturar interacciones semanticas cruzadas complejas, mejorando los margenes de precision frente a GAT estatico.
- **Contraste con el Baseline MLP:** Al entrenar un clasificador sin informacion de la estructura del grafo, el modelo se vera forzado a predecir la categoria basandose de manera restrictiva en la presencia aislada de palabras clave. Esto resultara en una degradacion notable de las metricas de precision y F1-Score en comparacion con los clasificadores basados en grafos, evidenciando experimentalmente el valor anadido de explotar la topologia de similitud textual.

### Caso 2: Clasificacion de noticias con grafo de similitud (AG News)

Este experimento introduce una metodologia hibrida donde la red de relaciones no existe de manera natural en los datos crudos, sino que se construye artificialmente estimando la similitud coseno de los vectores de frecuencia de palabras TF-IDF del conjunto de datos y enlazando los textos mas parecidos mediante un algoritmo KNN.

- **Comportamiento esperado de GraphSAGE:** Representa la eleccion optima para liderar este experimento dinamico. Dado que en aplicaciones reales de procesamiento de lenguaje natural (NLP) ingresan de forma continua nuevos articulos de noticias, un modelo transductivo como GCN fallaria al no poder realizar inferencias sobre documentos externos sin ejecutar un proceso de re-entrenamiento destructivo y costoso sobre todo el corpus aumentado. GraphSAGE resuelve esta limitacion estructural: al ingresar un nuevo texto, se preprocesa su representacion TF-IDF, se localiza su vecindario de mayor similitud coseno en el grafo y se alimenta a traves de sus funciones de agregacion entrenadas y su paso de combinacion para derivar su embedding final y predecir su categoria de forma inductiva en milisegundos. El muestreo controlado uniformemente garantiza que las demandas de memoria permanezcan constantes e independientes del crecimiento progresivo del grafo de similitud del corpus.

## 8. Comparativa breve con otras arquitecturas de Redes Neuronales

Para entender el posicionamiento de las GNN en SentinelGrafo, es necesario contrastarlas con los paradigmas tradicionales:

**Frente a Perceptrones Multicapa (MLP / Redes Dense):** Un MLP procesa cada vector de texto de forma aislada bajo el supuesto matematico de que las observaciones son independientes e identicamente distribuidas (i.i.d.). No puede integrar la estructura del entorno. Intentar pasarle un grafo forzaria a aplanar matrices de adyacencia, rompiendo la invariancia ante permutaciones y provocando una explosion inmanejable de parametros.

**Frente a Redes Neuronales Convolucionales (CNN):** Las CNN tradicionales operan sobre dominios euclidianos bidimensionales rigidos (como rejillas de pixeles en imagenes) con vecindarios de tamano fijo y simetrico. Las GNN (especificamente GCN) generalizan esta operacion convolucional a dominios no euclidianos con geometrias irregulares y grafos arbitrarios de tamano variable.

**Frente a Redes Recurrentes (RNN / LSTM):** Las RNN y LSTM procesan informacion de forma estrictamente secuencial y unidireccional (tiempo o secuencias de palabras). Aunque GraphSAGE puede usar una celda LSTM como mecanismo de agregacion local, lo hace aplicando permutaciones aleatorias para eliminar el sesgo posicional, priorizando la estructura vecinal relacional por encima del orden lineal estricto.

## 3. Pipeline común — Limpieza de texto y construcción de grafos

Ambos casos de estudio comparten el mismo pipeline de preprocesamiento.
Se definen dos funciones reutilizables:

- **`clean_text(text)`** — Normaliza el texto: minúsculas, elimina URLs,
  HTML, puntuación y espacios sobrantes.
- **`build_graph(texts, labels, max_features, k)`** — Vectoriza con TF-IDF,
  construye el grafo con similitud coseno y k vecinos más cercanos,
  y empaqueta todo en un objeto `torch_geometric.data.Data` listo para la GNN.

```python
def clean_text(text):
    '''Limpia y normaliza un texto para el pipeline de NLP.'''
    if not isinstance(text, str):
        return ''
    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^a-zñáéíóúü\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text


def build_graph(texts, labels, max_features=3000, k=5):
    '''
    Construye un grafo textual a partir de una lista de textos.
    Retorna: (torch_geometric.data.Data, TfidfVectorizer, np.array TF-IDF)
    '''
    print(f'Vectorizando {len(texts)} textos con TF-IDF (max_features={max_features})...')
    vectorizer = TfidfVectorizer(max_features=max_features, stop_words='english')
    X_tfidf = vectorizer.fit_transform(texts).toarray()
    print(f'Matriz TF-IDF: {X_tfidf.shape}')

    from sklearn.neighbors import NearestNeighbors
    print(f'Buscando {k} vecinos mas cercanos con NearestNeighbors (coseno)...')
    knn = NearestNeighbors(n_neighbors=k+1, metric='cosine', n_jobs=-1)
    knn.fit(X_tfidf)
    _, indices = knn.kneighbors(X_tfidf)
    neighbor_indices = indices[:, 1:]  # excluye self (posicion 0)

    n = len(texts)
    edge_src = np.repeat(np.arange(n), k)
    edge_dst = neighbor_indices.flatten()

    edge_index = torch.tensor([edge_src, edge_dst], dtype=torch.long)
    x = torch.tensor(X_tfidf, dtype=torch.float)
    y = torch.tensor(labels, dtype=torch.long)

    data = Data(x=x, edge_index=edge_index, y=y)
    print(f'Grafo construido: {data.num_nodes} nodos, {data.num_edges} aristas, '
          f'{data.num_features} features/nodo, {len(torch.unique(y))} clases')

    return data, vectorizer, X_tfidf
```

Se realiza balanceo de clases, limpieza de texto y split 80/20 para train/test.

```python
print('=' * 60)
print('CASO 1: DETECCION DE TWEETS SOBRE DESASTRES REALES')
print('=' * 60)

df_tweets = pd.read_csv(RUTA_DISASTER_TRAIN)
print(f'Tweets cargados: {len(df_tweets)}')
print(f'Columnas: {list(df_tweets.columns)}')
print(f'Distribucion de clases:\n{df_tweets["target"].value_counts()}')

df_tweets['clean_text'] = df_tweets['text'].apply(clean_text)
df_tweets = df_tweets[df_tweets['clean_text'].str.len() > 0].reset_index(drop=True)

min_class = df_tweets['target'].value_counts().min()
df_c1 = pd.concat([
    df_tweets[df_tweets['target'] == 0].sample(n=min_class, random_state=42),
    df_tweets[df_tweets['target'] == 1].sample(n=min_class, random_state=42)
], ignore_index=True)
df_c1 = df_c1.sample(frac=1, random_state=42).reset_index(drop=True)
print(f'\nDataset balanceado Caso 1: {len(df_c1)} muestras')
print(f'Distribucion: {dict(df_c1["target"].value_counts())}')

texts_c1  = df_c1['clean_text'].tolist()
labels_c1 = df_c1['target'].tolist()

data_c1, vec_c1, tfidf_c1 = build_graph(texts_c1, labels_c1, max_features=3000, k=5)

idx_c1 = np.arange(len(df_c1))
train_idx_c1, test_idx_c1 = train_test_split(idx_c1, test_size=0.2,
                                            random_state=42, stratify=labels_c1)
train_mask_c1 = torch.zeros(data_c1.num_nodes, dtype=torch.bool)
test_mask_c1  = torch.zeros(data_c1.num_nodes, dtype=torch.bool)
train_mask_c1[train_idx_c1] = True
test_mask_c1[test_idx_c1]   = True
print(f'\nSplit Caso 1 — Train: {train_mask_c1.sum().item()}, Test: {test_mask_c1.sum().item()}')
```

## 6. Visualización de la topología de los grafos

Se visualiza un subconjunto de 800 nodos de cada grafo para apreciar la estructura
de conexiones generada por la similitud textual.

**Caso 1 (Disaster Tweets):** Naranja = Desastre real, Azul = No desastre
**Caso 2 (AG News):** 4 colores para las 4 categorias (World, Sports, Business, Sci/Tech)


```python
def plot_graph_subset(data, labels_tensor, title, colors, labels,
                      max_nodes=800, seed=42):
    '''Visualiza un subgrafo cohesionado de max_nodes seleccionado por BFS para mostrar la estructura real.'''
    from collections import defaultdict

    # 1. Construir lista de adyacencia del grafo completo
    edge_index = data.edge_index
    edge_src = edge_index[0].cpu().numpy()
    edge_dst = edge_index[1].cpu().numpy()

    adj = defaultdict(list)
    for u, v in zip(edge_src, edge_dst):
        adj[u].append(v)
        adj[v].append(u)

    # 2. Seleccionar nodos usando BFS para garantizar que estén conectados
    start_node = max(adj.keys(), key=lambda x: len(adj[x]))

    selected_nodes = set()
    queue = [start_node]

    random.seed(seed)

    while queue and len(selected_nodes) < max_nodes:
        curr = queue.pop(0)
        if curr not in selected_nodes:
            selected_nodes.add(curr)
            for neighbor in adj[curr]:
                if neighbor not in selected_nodes and neighbor not in queue:
                    queue.append(neighbor)

        # Si la componente se agota, tomamos otra semilla del residuo
        if not queue and len(selected_nodes) < max_nodes:
            remaining = set(range(data.num_nodes)) - selected_nodes
            if remaining:
                remaining_adj = {n: adj[n] for n in remaining if n in adj}
                if remaining_adj:
                    next_seed = max(remaining_adj.keys(), key=lambda x: len(remaining_adj[x]))
                else:
                    next_seed = random.choice(list(remaining))
                queue.append(next_seed)
            else:
                break

    # 3. Crear el subgrafo inducido por los nodos seleccionados
    node_list = sorted(list(selected_nodes))
    node_to_idx = {old_id: new_id for new_id, old_id in enumerate(node_list)}

    sub_edges_src = []
    sub_edges_dst = []
    node_set = set(node_list)
    for u in node_list:
        for v in adj[u]:
            if v in node_set and u < v:
                sub_edges_src.append(node_to_idx[u])
                sub_edges_dst.append(node_to_idx[v])

    sub_ei = torch.tensor([sub_edges_src, sub_edges_dst], dtype=torch.long)
    sub_x = data.x[node_list]

    if torch.is_tensor(labels_tensor):
        y_np = labels_tensor.cpu().numpy()
    else:
        y_np = np.array(labels_tensor)

    sub_y = torch.tensor(y_np[node_list], dtype=torch.long)
    sub_data = Data(x=sub_x, edge_index=sub_ei, y=sub_y)
    G = to_networkx(sub_data, to_undirected=True)

    node_colors = [colors[lbl] for lbl in y_np[node_list]]

    # 4. Dibujar el grafo de forma atractiva
    fig, ax = plt.subplots(figsize=(12, 10))
    pos = nx.spring_layout(G, seed=seed, k=0.08, iterations=100)

    # Aristas con opacidad suave
    nx.draw_networkx_edges(G, pos, alpha=0.12, edge_color='gray', ax=ax)
    # Nodos
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=35,
                           alpha=0.9, ax=ax)

    legend_elements = [mpatches.Patch(color=colors[i], label=labels[i])
                       for i in range(len(colors))]
    ax.legend(handles=legend_elements, loc='upper right', fontsize=10, frameon=True)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.axis('off')
    plt.tight_layout()
    plt.show()


print('\n>>> VISUALIZACION 1: TOPOLOGIA DE LOS GRAFOS <<<\n')

plot_graph_subset(data_c1, labels_c1,
                  'Caso 1: Grafo de Disaster Tweets (muestra de 800 nodos)',
                  colors=['#FF8C00', '#1E90FF'],
                  labels=['Desastre real', 'No desastre'])

plot_graph_subset(data_c2, labels_c2,
                  'Caso 2: Grafo de AG News (muestra de 800 nodos)',
                  colors=['#3498DB', '#E74C3C', '#2ECC71', '#F39C12'],
                  labels=['World', 'Sports', 'Business', 'Sci/Tech'])
```

## 7. Baseline: Logistic Regression sobre vectores TF-IDF

Antes de entrenar las GNN, se establece un baseline con Regresión Logística
entrenada directamente sobre los vectores TF-IDF (sin información del grafo).

Esto permite medir cuánto aporta la estructura de grafo y el message passing de la GNN.

```python
def run_baseline(tfidf_matrix, labels, train_idx, test_idx, case_name):
    '''Entrena Logistic Regression sobre TF-IDF y retorna metricas.'''
    X_train = tfidf_matrix[train_idx]
    X_test  = tfidf_matrix[test_idx]
    y_train = np.array(labels)[train_idx]
    y_test  = np.array(labels)[test_idx]

    model = LogisticRegression(max_iter=2000, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    acc  = accuracy_score(y_test, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='macro')

    print(f'--- Baseline Logistic Regression: {case_name} ---')
    print(f'Accuracy:  {acc:.4f}')
    print(f'Precision: {prec:.4f}')
    print(f'Recall:    {rec:.4f}')
    print(f'F1-score:  {f1:.4f}')
    target_names = ['Clase 0', 'Clase 1'] if len(set(labels)) == 2 else ['World', 'Sports', 'Business', 'Sci/Tech']
    print(classification_report(y_test, y_pred,
          target_names=target_names))

    return {'model': model, 'acc': acc, 'prec': prec, 'rec': rec, 'f1': f1,
            'y_pred': y_pred, 'y_true': y_test}


print('\n>>> EVALUACION BASELINE <<<\n')

baseline_c1 = run_baseline(tfidf_c1, labels_c1, train_idx_c1, test_idx_c1,
                           'Caso 1: Disaster Tweets')
print()
baseline_c2 = run_baseline(tfidf_c2, labels_c2, train_idx_c2, test_idx_c2,
                           'Caso 2: AG News')
```

## 8. Definición de modelos GNN: GraphSAGE y GCN

Se implementa una clase `GNNClassifier` unificada que acepta un parámetro `model_type`
para seleccionar la arquitectura:

- **GraphSAGE (`sage`)**: Utiliza `SAGEConv` de PyG. Aprende una función de agregación
  sobre los vecinos muestreados. Es inductivo: puede generalizar a nodos no vistos.
- **GCN (`gcn`)**: Utiliza `GCNConv` de PyG. Aplica una convolución espectral
  de primer orden sobre todo el grafo.

Ambos modelos comparten la misma estructura: 2 capas de convolución de grafo con
activación ReLU, dropout para regularización, y una capa lineal final de clasificación.

El método `get_embeddings()` extrae las representaciones latentes de la penúltima capa,
útiles para la visualización t-SNE.

```python
class GNNClassifier(nn.Module):
    '''
    Clasificador basado en Graph Neural Networks.
    Soporta GraphSAGE y GCN como backbones.
    '''
    def __init__(self, in_channels, hidden_channels, out_channels,
                 model_type='sage', dropout=0.5):
        super().__init__()
        self.model_type = model_type
        self.dropout = dropout
        self.hidden_channels = hidden_channels

        if model_type == 'sage':
            self.conv1 = SAGEConv(in_channels, hidden_channels)
            self.conv2 = SAGEConv(hidden_channels, hidden_channels)
        elif model_type == 'gcn':
            self.conv1 = GCNConv(in_channels, hidden_channels)
            self.conv2 = GCNConv(hidden_channels, hidden_channels)
        else:
            raise ValueError(f"Modelo '{model_type}' no soportado. Use 'sage' o 'gcn'.")

        self.classifier = nn.Linear(hidden_channels, out_channels)
        self._embeddings = None

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)
        x = self.conv2(x, edge_index)
        x = F.relu(x)
        self._embeddings = x
        x = F.dropout(x, p=self.dropout, training=self.training)
        x = self.classifier(x)
        return x

    def get_embeddings(self):
        '''Retorna las embeddings de la penultima capa.'''
        return self._embeddings.detach().cpu().numpy()


print('Clase GNNClassifier definida correctamente.')
print('Arquitecturas soportadas: GraphSAGE (sage), GCN (gcn)')
```

## 9. Función unificada de entrenamiento y evaluación

`train_and_evaluate` entrena un modelo GNN con:
- Optimizador Adam con weight decay para regularización L2
- Cross-entropy loss
- Early stopping con paciencia configurable (restaura los mejores pesos)
- Registro de pérdida y accuracy por época para curvas de aprendizaje
- Extracción de embeddings finales para visualización t-SNE

```python
def train_and_evaluate(data, train_mask, test_mask, model_type='sage',
                       hidden_dim=64, epochs=200, lr=0.01, patience=30):
    '''
    Entrena un GNNClassifier y retorna modelo, historial, predicciones,
    embeddings y mejor accuracy en test.
    '''
    model = GNNClassifier(
        in_channels=data.num_features,
        hidden_channels=hidden_dim,
        out_channels=len(torch.unique(data.y)),
        model_type=model_type,
        dropout=0.5
    ).to(DEVICE)

    data = data.to(DEVICE)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=5e-4)
    criterion = nn.CrossEntropyLoss()

    history = {'train_loss': [], 'train_acc': [], 'test_acc': [], 'test_loss': []}
    best_test_acc = 0.0
    best_state = None
    patience_counter = 0

    pbar_prefix = f'{model_type.upper()} '
    for epoch in range(epochs):
        model.train()
        optimizer.zero_grad()
        out = model(data.x, data.edge_index)
        loss = criterion(out[train_mask], data.y[train_mask])
        loss.backward()
        optimizer.step()

        model.eval()
        with torch.no_grad():
            out_eval = model(data.x, data.edge_index)
            train_pred = out_eval[train_mask].argmax(dim=1)
            train_acc = (train_pred == data.y[train_mask]).float().mean().item()
            test_pred = out_eval[test_mask].argmax(dim=1)
            test_acc = (test_pred == data.y[test_mask]).float().mean().item()
            test_loss_val = criterion(out_eval[test_mask], data.y[test_mask]).item()

        history['train_loss'].append(loss.item())
        history['train_acc'].append(train_acc)
        history['test_acc'].append(test_acc)
        history['test_loss'].append(test_loss_val)

        if test_acc > best_test_acc:
            best_test_acc = test_acc
            best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
            patience_counter = 0
        else:
            patience_counter += 1

        if patience_counter >= patience:
            print(f'{pbar_prefix}Early stopping en epoca {epoch+1}/{epochs}')
            break
    else:
        print(f'{pbar_prefix}Entrenamiento completo: {epochs} epocas')

    if best_state is not None:
        model.load_state_dict(best_state)
    model.eval()

    with torch.no_grad():
        out = model(data.x, data.edge_index)
        test_pred = out[test_mask].argmax(dim=1).cpu().numpy()
        test_true = data.y[test_mask].cpu().numpy()
        embeddings = model.get_embeddings()

    results = {
        'model': model,
        'history': history,
        'y_pred': test_pred,
        'y_true': test_true,
        'embeddings': embeddings,
        'best_acc': best_test_acc,
    }
    return results


print('Funcion train_and_evaluate lista.')
```

## 13. Visualización 3 — Proyección t-SNE de embeddings

Se extraen los embeddings de la penúltima capa de cada GNN (64 dimensiones)
y se proyectan a 2D usando t-SNE.

Esto permite observar si el modelo logra separar las clases en el espacio latente.
Los colores indican la clase real; los marcadores, la predicción del modelo.
Un círculo alrededor de un punto indica que el modelo acertó; una X indica error.

```python
print('\n>>> VISUALIZACION 4: PROYECCION t-SNE DE EMBEDDINGS <<<\n')

def plot_tsne(embeddings, y_true, y_pred, title, ax, colors, class_names):
    '''Proyecta embeddings con t-SNE y colorea por clase real + acierto/error.'''
    tsne = TSNE(n_components=2, random_state=42, perplexity=min(30, len(embeddings)//3))
    emb_2d = tsne.fit_transform(embeddings)

    correct = (y_true == y_pred)
    for cls_val in range(len(colors)):
        mask_correct = (y_true == cls_val) & correct
        mask_error = (y_true == cls_val) & ~correct
        if mask_correct.any():
            ax.scatter(emb_2d[mask_correct, 0], emb_2d[mask_correct, 1],
                       c=colors[cls_val], marker='o', s=30, alpha=0.7,
                       edgecolors='white', linewidth=0.3,
                       label=f'{class_names[cls_val]} (OK)')
        if mask_error.any():
            ax.scatter(emb_2d[mask_error, 0], emb_2d[mask_error, 1],
                       c=colors[cls_val], marker='X', s=60, alpha=0.9,
                       edgecolors='black', linewidth=0.5,
                       label=f'{class_names[cls_val]} (Error)')

    ax.set_title(title, fontsize=11, fontweight='bold')
    ax.set_xlabel('t-SNE 1')
    ax.set_ylabel('t-SNE 2')
    ax.legend(fontsize=6, loc='best')


fig, axes = plt.subplots(2, 2, figsize=(18, 16))

tsne_configs = [
    (0, 0, 'C1_GraphSAGE', 'Caso 1 — GraphSAGE Embeddings',
     ['#FF8C00', '#1E90FF'], ['Desastre', 'No Desastre']),
    (0, 1, 'C1_GCN',       'Caso 1 — GCN Embeddings',
     ['#FF8C00', '#1E90FF'], ['Desastre', 'No Desastre']),
    (1, 0, 'C2_GraphSAGE', 'Caso 2 — GraphSAGE Embeddings',
     ['#3498DB', '#E74C3C', '#2ECC71', '#F39C12'],
     ['World', 'Sports', 'Business', 'Sci/Tech']),
    (1, 1, 'C2_GCN',       'Caso 2 — GCN Embeddings',
     ['#3498DB', '#E74C3C', '#2ECC71', '#F39C12'],
     ['World', 'Sports', 'Business', 'Sci/Tech']),
]

for row, col, key, title, colors, class_names in tsne_configs:
    r = all_results[key]
    mask_test = test_mask_c1.numpy() if 'C1' in key else test_mask_c2.numpy()
    plot_tsne(r['embeddings'][mask_test], r['y_true'], r['y_pred'],
              title, axes[row, col], colors, class_names)

plt.suptitle('Visualizacion t-SNE de Embeddings de las GNN',
             fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

## 14. Visualización 4 — Diagrama esquemático de capas GNN (Message Passing)

Se dibuja un diagrama conceptual del mecanismo de **Message Passing** que usan
GraphSAGE y GCN. El flujo es:

1. **Nodos de entrada** con features (TF-IDF, 3000 dims)
2. **AGGREGATE**: cada nodo recibe mensajes de sus vecinos
3. **UPDATE**: se combina la información propia con la agregada (MLP)
4. **Nueva representación** (embedding, 64 dims)
5. **Clasificador**: capa lineal que produce logits para cada clase

Esto se repite en 2 capas de convolución para propagar información a 2 saltos de distancia.

```python
print('\n>>> VISUALIZACION 5: DIAGRAMA DE CAPAS GNN (MESSAGE PASSING) <<<\n')

def draw_gnn_diagram():
    fig, ax = plt.subplots(1, 1, figsize=(16, 7))
    ax.set_xlim(0, 16)
    ax.set_ylim(0, 7)
    ax.axis('off')

    def draw_box(x, y, w, h, text, color, fontsize=10, text_color='white'):
        rect = mpatches.FancyBboxPatch((x, y), w, h, boxstyle='round,pad=0.15',
                                       facecolor=color, edgecolor='black',
                                       linewidth=1.5, alpha=0.9)
        ax.add_patch(rect)
        ax.text(x + w/2, y + h/2, text, ha='center', va='center',
                fontsize=fontsize, fontweight='bold', color=text_color)

    def draw_arrow(x1, y1, x2, y2, label='', color='gray'):
        ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                    arrowprops=dict(arrowstyle='->', color=color, lw=2,
                                    connectionstyle='arc3,rad=0'))
        if label:
            mx, my = (x1 + x2) / 2, (y1 + y2) / 2 + 0.25
            ax.text(mx, my, label, ha='center', va='bottom',
                    fontsize=9, fontstyle='italic', color=color)

    y_center = 3.5

    draw_box(0.5, y_center-0.8, 2.8, 1.6,
             'NODOS DE\nENTRADA\n(features TF-IDF)', '#2C3E50', 10)

    draw_box(5.0, y_center-0.8, 2.8, 1.6,
             'AGGREGATE\n(Mensajes de\nvecinos k-NN)', '#2980B9', 10)

    draw_box(9.0, y_center-0.8, 2.8, 1.6,
             'UPDATE\n(MLP + ReLU\n+ Dropout)', '#27AE60', 10)

    draw_box(13.0, y_center-0.8, 2.8, 1.6,
             'EMBEDDINGS\n(64 dimensiones\nespacio latente)', '#8E44AD', 10)

    draw_box(6.5, 0.8, 4.5, 1.2,
             'CLASIFICADOR\nCapa Lineal -> Logits (2 clases)', '#E74C3C', 10)

    draw_arrow(3.3, y_center, 5.0, y_center, 'message', '#3498DB')
    draw_arrow(7.8, y_center, 9.0, y_center, 'passing', '#2ECC71')
    draw_arrow(11.8, y_center, 13.0, y_center, '', '#9B59B6')

    ax.annotate('', xy=(8.75, 1.8), xytext=(8.75, y_center-0.8),
                arrowprops=dict(arrowstyle='->', color='#7F8C8D', lw=2,
                                connectionstyle='arc3,rad=0'))

    ax.annotate('', xy=(8.75, 0.6), xytext=(8.75, 2.0),
                arrowprops=dict(arrowstyle='->', color='#7F8C8D', lw=2,
                                connectionstyle='arc3,rad=0.3'))
    ax.text(9.3, 1.3, 'x2 capas\n(2-hop\nmessage\npassing)',
            ha='left', va='center', fontsize=8, fontstyle='italic', color='#7F8C8D')

    ax.text(8, 6.3, 'DIAGRAMA DE MESSAGE PASSING EN GNN (GraphSAGE / GCN)',
            ha='center', va='center', fontsize=16, fontweight='bold', color='#2C3E50')
    ax.text(8, 5.6, 'Cada nodo agrega informacion de sus vecinos y actualiza su representacion',
            ha='center', va='center', fontsize=11, color='#7F8C8D')

    plt.tight_layout()
    plt.show()

draw_gnn_diagram()
```

**Cómo funciona:** El nuevo texto se limpia, se vectoriza con TF-IDF,
se conecta al grafo existente mediante k-NN, y se ejecuta el forward pass
de la GNN para obtener la predicción del nuevo nodo.


```python
# ===================================================================
# FUNCIONES DE AUXILIO E INFERENCIA
# ===================================================================
def predict_new_text(raw_text, model, data, vectorizer, tfidf_matrix, labels,
                     texts_original, k=5):
    '''Predice la clase de un nuevo texto usando la GNN entrenada.'''
    clean = clean_text(raw_text)
    if len(clean) < 3:
        return {
            'ok': False,
            'message': 'El texto es demasiado corto. Ingresa al menos 3 caracteres.'
        }

    new_vec = vectorizer.transform([clean]).toarray()
    sims = cosine_similarity(new_vec, tfidf_matrix)[0]
    neighbor_idx = np.argsort(sims)[-k:][::-1]

    new_x = torch.tensor(new_vec, dtype=torch.float)
    num_hops = get_message_passing_hops(model)
    local_x, local_edge_index, local_target_idx, local_context = build_local_prediction_subgraph(
        data, new_x, neighbor_idx, num_hops, labels
    )

    model.eval()
    with torch.no_grad():
        out = model(local_x.to(DEVICE), local_edge_index.to(DEVICE))
        probs = F.softmax(out[local_target_idx], dim=0).cpu().numpy()
        pred_class = int(probs.argmax())
        confidence = float(probs.max())

    neighbors_info = []
    for idx in neighbor_idx:
        similarity = float(sims[idx])
        label_str = str(labels[idx])
        snippet = texts_original[idx][:100] + '...' if len(texts_original[idx]) > 100 else texts_original[idx]
        neighbors_info.append({
            'dataset_idx': int(idx),
            'similarity': similarity,
            'label': label_str,
            'snippet': snippet
        })

    inference_meta = {
        'num_hops': num_hops,
        'subgraph_nodes': int(local_context['num_nodes']),
        'anchor_neighbors': int(len(neighbor_idx))
    }
    return {
        'ok': True,
        'pred_class': pred_class,
        'confidence': confidence,
        'probabilities': probs,
        'neighbors_info': neighbors_info,
        'inference_meta': inference_meta,
        'local_context': local_context
    }


def infer_case_data(case_key):
    if case_key == 'c1':
        return (all_results['C1_GraphSAGE']['model'], data_c1, vec_c1, tfidf_c1,
                np.array(labels_c1), df_c1['text'].tolist())
    return (all_results['C2_GraphSAGE']['model'], data_c2, vec_c2, tfidf_c2,
            np.array(labels_c2), df_c2['content'].tolist())


def get_case_visual_config(case_key):
    if case_key == 'c1':
        class_names = {0: 'NO DESASTRE', 1: 'DESASTRE REAL'}
        button_styles = {0: 'info', 1: 'danger'}
        border_colors = {0: '#3498DB', 1: '#E74C3C'}
    else:
        class_names = {0: 'World', 1: 'Sports', 2: 'Business', 3: 'Sci/Tech'}
        button_styles = {0: 'info', 1: 'danger', 2: 'success', 3: 'warning'}
        border_colors = {0: '#3498DB', 1: '#E74C3C', 2: '#2ECC71', 3: '#F39C12'}
    return class_names, button_styles, border_colors


def get_message_passing_hops(model):
    hops = sum(1 for module in model.modules() if isinstance(module, (SAGEConv, GCNConv)))
    return max(1, hops)


def build_local_prediction_subgraph(data, new_x, neighbor_idx, num_hops, labels):
    neighbor_idx = np.array(neighbor_idx, dtype=int)
    anchor_set = set(neighbor_idx.tolist())
    new_node_idx = int(data.num_nodes)

    bridge_src = [new_node_idx] * len(neighbor_idx) + neighbor_idx.tolist()
    bridge_dst = neighbor_idx.tolist() + [new_node_idx] * len(neighbor_idx)
    bridge_edges = torch.tensor([bridge_src, bridge_dst], dtype=torch.long)
    augmented_edge_index = torch.cat([data.edge_index.cpu(), bridge_edges], dim=1)

    subset, sub_edge_index, mapping, _ = k_hop_subgraph(
        new_node_idx,
        num_hops,
        augmented_edge_index,
        relabel_nodes=True
    )

    base_x = data.x.cpu()
    local_rows = []
    local_labels = []
    anchor_local_indices = []
    original_indices = subset.tolist()

    for local_idx, node_idx in enumerate(original_indices):
        if node_idx == new_node_idx:
            local_rows.append(new_x[0])
            local_labels.append(-1)
        else:
            local_rows.append(base_x[node_idx])
            local_labels.append(int(labels[node_idx]))
            if node_idx in anchor_set:
                anchor_local_indices.append(local_idx)

    local_x = torch.stack(local_rows, dim=0)
    local_context = {
        'num_nodes': int(len(original_indices)),
        'original_indices': original_indices,
        'local_labels': local_labels,
        'anchor_local_indices': anchor_local_indices,
        'target_local_idx': int(mapping.item()),
        'local_edge_index': sub_edge_index.cpu()
    }
    return local_x, sub_edge_index, int(mapping.item()), local_context


def build_probability_widget(class_names, probs, border_colors, pred_class):
    ordered_classes = list(class_names.keys())
    labels_ordered = [class_names[idx] for idx in ordered_classes]
    values = [float(probs[idx]) for idx in ordered_classes]
    colors = [border_colors[idx] for idx in ordered_classes]

    output = widgets.Output(layout=widgets.Layout(border='1px solid #D9E2EC', padding='6px', width='100%'))
    with output:
        fig, ax = plt.subplots(figsize=(6.8, 3.0))
        ax.set_facecolor('#FBFDFF')
        bars = ax.barh(labels_ordered, values, color=colors, alpha=0.88)
        for cls_idx, bar in zip(ordered_classes, bars):
            if cls_idx == pred_class:
                bar.set_edgecolor('#1F2937')
                bar.set_linewidth(2.2)
            else:
                bar.set_edgecolor('white')
                bar.set_linewidth(1.0)
            ax.text(
                min(bar.get_width() + 0.015, 1.02),
                bar.get_y() + bar.get_height() / 2,
                f'{bar.get_width():.1%}',
                va='center',
                fontsize=10
            )

        ax.set_xlim(0, 1.05)
        ax.set_xlabel('Probabilidad')
        ax.set_title('Distribucion de probabilidades', fontsize=12, pad=10)
        ax.grid(axis='x', linestyle='--', alpha=0.25)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.spines['bottom'].set_alpha(0.35)
        plt.tight_layout()
        display(fig)
        plt.close(fig)
    return output


def build_prediction_graph_widget(case_key, prediction_result):
    class_names, _, border_colors = get_case_visual_config(case_key)
    pred_class = prediction_result['pred_class']
    local_context = prediction_result['local_context']
    edge_index = local_context['local_edge_index'].numpy()
    target_idx = local_context['target_local_idx']
    anchor_local = set(local_context['anchor_local_indices'])
    local_labels = local_context['local_labels']

    G = nx.Graph()
    G.add_nodes_from(range(local_context['num_nodes']))
    G.add_edges_from(zip(edge_index[0], edge_index[1]))

    if G.number_of_nodes() == 0:
        return widgets.Label(value='No se pudo construir el subgrafo local de la prediccion.')

    pos = nx.spring_layout(
        G,
        seed=42,
        k=1.4 / np.sqrt(max(G.number_of_nodes(), 1)),
        iterations=80
    )
    tx, ty = pos[target_idx]
    pos = {node: (coords[0] - tx, coords[1] - ty) for node, coords in pos.items()}

    output = widgets.Output(layout=widgets.Layout(border='1px solid #D9E2EC', padding='6px', width='100%'))
    with output:
        fig, ax = plt.subplots(figsize=(7.8, 5.8))
        fig.patch.set_facecolor('white')
        ax.set_facecolor('#F8FBFF')

        highlighted_edges = []
        normal_edges = []
        for u, v in G.edges():
            if target_idx in (u, v):
                highlighted_edges.append((u, v))
            else:
                normal_edges.append((u, v))

        if normal_edges:
            nx.draw_networkx_edges(
                G, pos, edgelist=normal_edges, ax=ax,
                edge_color='#B9C3CF', alpha=0.35, width=1.0
            )
        if highlighted_edges:
            nx.draw_networkx_edges(
                G, pos, edgelist=highlighted_edges, ax=ax,
                edge_color=border_colors[pred_class], alpha=0.75, width=2.6
            )

        for cls_idx, color in border_colors.items():
            nodes = [i for i, lbl in enumerate(local_labels) if lbl == cls_idx and i != target_idx]
            if not nodes:
                continue
            node_sizes = [280 if i in anchor_local else 135 for i in nodes]
            line_widths = [2.0 if i in anchor_local else 0.9 for i in nodes]
            edge_colors = ['#1F2937' if i in anchor_local else 'white' for i in nodes]
            nx.draw_networkx_nodes(
                G, pos, nodelist=nodes, ax=ax,
                node_color=color,
                node_size=node_sizes,
                alpha=0.92,
                edgecolors=edge_colors,
                linewidths=line_widths
            )

        nx.draw_networkx_nodes(
            G, pos, nodelist=[target_idx], ax=ax,
            node_color=border_colors[pred_class],
            node_size=560,
            node_shape='*',
            edgecolors='#111827',
            linewidths=1.8
        )

        ax.text(
            pos[target_idx][0],
            pos[target_idx][1] + 0.10,
            'Texto nuevo',
            ha='center',
            va='bottom',
            fontsize=11,
            fontweight='bold',
            color='#111827'
        )

        for local_idx in list(anchor_local)[:3]:
            ax.text(
                pos[local_idx][0],
                pos[local_idx][1] - 0.08,
                'vecino ancla',
                ha='center',
                va='top',
                fontsize=8.5,
                color='#374151'
            )

        ax.set_title('Subgrafo local usado en la prediccion', fontsize=13, pad=12)
        ax.text(
            0.02, 0.02,
            'Estrella = texto nuevo | Nodos mas grandes = vecinos ancla',
            transform=ax.transAxes,
            fontsize=9,
            color='#4B5563',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='#D1D5DB', alpha=0.9)
        )
        ax.axis('off')
        plt.tight_layout()
        display(fig)
        plt.close(fig)
    return output


def render_prediction_report(case_key, prediction_result):
    class_names, button_styles, border_colors = get_case_visual_config(case_key)
    pred_class = prediction_result['pred_class']
    confidence = prediction_result['confidence']
    inference_meta = prediction_result['inference_meta']
    neighbors_info = prediction_result['neighbors_info']

    explanation_text = (
        f'Se vectorizo el texto, se conecto con {inference_meta["anchor_neighbors"]} vecinos mas similares, '
        f'se extrajo un subgrafo local de {inference_meta["num_hops"]} saltos '
        f'con {inference_meta["subgraph_nodes"]} nodos y luego GraphSAGE realizo message passing para clasificarlo.'
    )

    main_color = border_colors[pred_class]
    result_card = widgets.VBox([
        widgets.Label(value='Resultado'),
        widgets.Button(
            description=class_names[pred_class],
            disabled=True,
            button_style=button_styles.get(pred_class, ''),
            layout=widgets.Layout(width='260px', height='38px')
        ),
        widgets.Label(value=f'Confianza: {confidence:.1%}'),
        widgets.Label(
            value=(
                f"Inferencia GNN local: {inference_meta['subgraph_nodes']} nodos | "
                f"{inference_meta['num_hops']} saltos | "
                f"{inference_meta['anchor_neighbors']} vecinos ancla"
            )
        )
    ], layout=widgets.Layout(
        border=f'2px solid {main_color}',
        padding='14px',
        margin='0 0 10px 0',
        width='100%'
    ))

    explanation_box = widgets.Box(
        [widgets.Label(value=explanation_text)],
        layout=widgets.Layout(
            border='1px solid #D9E2EC',
            padding='12px',
            margin='0 0 10px 0',
            width='100%'
        )
    )

    probability_widget = build_probability_widget(
        class_names,
        prediction_result['probabilities'],
        border_colors,
        pred_class
    )
    graph_widget = build_prediction_graph_widget(case_key, prediction_result)

    sections = [
        result_card,
        explanation_box,
        widgets.Label(value='Probabilidades por clase'),
        probability_widget,
        widgets.Label(value='Vecindario local utilizado por la GNN'),
        graph_widget
    ]

    if isinstance(neighbors_info, list) and neighbors_info:
        sections.append(widgets.Label(value='Vecinos mas similares en el grafo'))
        for i, neighbor in enumerate(neighbors_info, start=1):
            raw_label = neighbor['label']
            neighbor_label = int(raw_label) if str(raw_label).isdigit() else raw_label
            if isinstance(neighbor_label, int):
                neighbor_label_text = class_names.get(neighbor_label, str(neighbor_label))
                neighbor_color = border_colors.get(neighbor_label, '#999999')
            else:
                neighbor_label_text = str(neighbor_label)
                neighbor_color = '#999999'

            snippet_box = widgets.Textarea(
                value=neighbor['snippet'],
                disabled=True,
                layout=widgets.Layout(width='100%', height='72px')
            )
            neighbor_card = widgets.VBox([
                widgets.Label(value=f'Vecino #{i} | similitud: {neighbor["similarity"]:.3f}'),
                widgets.Label(value=f'Etiqueta: {neighbor_label_text}'),
                snippet_box
            ], layout=widgets.Layout(
                border=f'2px solid {neighbor_color}',
                padding='10px',
                margin='4px 0',
                width='100%'
            ))
            sections.append(neighbor_card)

    set_prediction_widgets(sections)


# Cerrar el panel anterior si la celda se ejecuta varias veces.
for widget_name in ['panel_title', 'tabs']:
    old_widget = globals().get(widget_name)
    if hasattr(old_widget, 'close'):
        old_widget.close()

panel_state = {
    'predict_busy': False,
    'graph_busy': False,
    'graph_cache': {}
}


# ===================================================================
# TAB 1: PREDICCION EN VIVO
# ===================================================================
case_selector = widgets.Dropdown(
    options=[
        ('Caso 1: Disaster Tweets', 'c1'),
        ('Caso 2: AG News', 'c2'),
    ],
    value='c1',
    description='Caso:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='300px')
)

text_input = widgets.Textarea(
    placeholder='Escribe un tweet o noticia en ingles para clasificar...',
    description='Texto:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='90%', height='100px')
)

predict_btn = widgets.Button(
    description='PREDECIR',
    button_style='primary',
    icon='search',
    layout=widgets.Layout(width='150px', height='40px')
)

output_area = widgets.VBox(
    [],
    layout=widgets.Layout(width='92%', min_height='70px', margin='8px 0 0 0')
)


def set_prediction_widgets(children):
    output_area.children = tuple(children)


def set_prediction_message(message, border_color='#D9D9D9'):
    message_box = widgets.Box(
        [widgets.Label(value=message)],
        layout=widgets.Layout(
            border=f'2px solid {border_color}',
            padding='12px',
            margin='0 0 8px 0',
            width='100%'
        )
    )
    set_prediction_widgets([message_box])


def on_predict_clicked(b):
    if panel_state['predict_busy']:
        return

    panel_state['predict_busy'] = True
    predict_btn.disabled = True

    try:
        set_prediction_widgets([])
        raw_text = text_input.value.strip()
        if not raw_text:
            set_prediction_message('Por favor ingresa un texto para clasificar.', border_color='#E74C3C')
            return

        model, data, vec, tfidf, labels, texts_orig = infer_case_data(case_selector.value)
        prediction_result = predict_new_text(
            raw_text, model, data, vec, tfidf, labels, texts_orig
        )

        if not prediction_result['ok']:
            set_prediction_message(prediction_result['message'], border_color='#E74C3C')
            return

        render_prediction_report(case_selector.value, prediction_result)
    except Exception as exc:
        set_prediction_message(f'Error durante la prediccion: {exc}', border_color='#E74C3C')
    finally:
        predict_btn.disabled = False
        panel_state['predict_busy'] = False


predict_btn.on_click(on_predict_clicked, remove=True)
predict_btn.on_click(on_predict_clicked)

set_prediction_message('Escribe un texto y pulsa PREDECIR para ver el resultado.', border_color='#D9D9D9')

pred_tab_content = widgets.VBox([
    case_selector,
    text_input,
    predict_btn,
    output_area
], layout=widgets.Layout(padding='10px'))


# ===================================================================
# TAB 2: EXPLORADOR DE GRAFOS (MATPLOTLIB)
# ===================================================================
graph_case_select = widgets.Dropdown(
    options=[
        ('Caso 1: Disaster Tweets', 'c1'),
        ('Caso 2: AG News (4 categorias)', 'c2'),
    ],
    value='c1',
    description='Caso:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='280px')
)

graph_nodes_slider = widgets.IntSlider(
    value=800, min=100, max=2000, step=100,
    description='Nodos:',
    continuous_update=False,
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='380px')
)

graph_k_slider = widgets.IntSlider(
    value=5, min=3, max=15, step=1,
    description='k-vecinos:',
    continuous_update=False,
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='320px')
)

node_size_slider = widgets.IntSlider(
    value=8, min=3, max=25, step=1,
    description='Tamano nodo:',
    continuous_update=False,
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='320px')
)

edge_opacity_slider = widgets.FloatSlider(
    value=0.15, min=0.0, max=0.5, step=0.05,
    description='Opacidad aristas:',
    continuous_update=False,
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='350px')
)

graph_animate_toggle = widgets.Checkbox(
    value=True,
    description='Animar render',
    indent=False,
    layout=widgets.Layout(width='150px')
)

graph_regen_btn = widgets.Button(
    description='Visualizar Grafo',
    button_style='info',
    icon='eye',
    layout=widgets.Layout(width='180px', height='36px')
)

graph_info_output = widgets.Output(layout=widgets.Layout(border='1px solid #D9D9D9', padding='8px'))
graph_output = widgets.Output()


def update_graph_info(message_lines):
    with graph_info_output:
        clear_output(wait=True)
        for line in message_lines:
            print(line)


def get_graph_source(case_key, k_neighbors):
    cache_key = (case_key, k_neighbors)
    if cache_key in panel_state['graph_cache']:
        return panel_state['graph_cache'][cache_key]

    if case_key == 'c1':
        data_src = data_c1
        labels_src = labels_c1
        colors = ['#FF8C00', '#1E90FF']
        class_labels = ['Desastre real', 'No desastre']
        title_base = 'Caso 1: Disaster Tweets'
        if k_neighbors != 5:
            data_src, _, _ = build_graph(texts_c1, labels_c1, max_features=3000, k=k_neighbors)
    else:
        data_src = data_c2
        labels_src = labels_c2
        colors = ['#3498DB', '#E74C3C', '#2ECC71', '#F39C12']
        class_labels = ['World', 'Sports', 'Business', 'Sci/Tech']
        title_base = 'Caso 2: AG News - 4 categorias'
        if k_neighbors != 5:
            data_src, _, _ = build_graph(texts_c2, labels_c2, max_features=5000, k=k_neighbors)

    payload = (data_src, np.array(labels_src), colors, class_labels, title_base)
    panel_state['graph_cache'][cache_key] = payload
    return payload


def build_connected_subgraph(data_src, labels_src, n_nodes):
    from collections import defaultdict, deque

    n = min(n_nodes, data_src.num_nodes)
    edge_index = data_src.edge_index.cpu().numpy()
    adj = defaultdict(set)
    for u, v in zip(edge_index[0], edge_index[1]):
        if u == v:
            continue
        adj[int(u)].add(int(v))
        adj[int(v)].add(int(u))

    if not adj:
        idx_sample = list(range(n))
    else:
        start_node = max(adj.keys(), key=lambda node: len(adj[node]))
        selected_nodes = set()
        queue = deque([start_node])

        while queue and len(selected_nodes) < n:
            curr = queue.popleft()
            if curr in selected_nodes:
                continue
            selected_nodes.add(curr)
            for neighbor in sorted(adj[curr]):
                if neighbor not in selected_nodes:
                    queue.append(neighbor)

            if not queue and len(selected_nodes) < n:
                remaining = [node for node in range(data_src.num_nodes) if node not in selected_nodes]
                if remaining:
                    next_seed = max(remaining, key=lambda node: len(adj[node]))
                    queue.append(next_seed)

        idx_sample = sorted(selected_nodes)

    node_to_idx = {old: new for new, old in enumerate(idx_sample)}
    node_set = set(idx_sample)
    sub_src = []
    sub_dst = []

    for old_u in idx_sample:
        for old_v in adj.get(old_u, []):
            if old_v in node_set and old_u < old_v:
                sub_src.append(node_to_idx[old_u])
                sub_dst.append(node_to_idx[old_v])

    sub_edge_index = torch.tensor([sub_src + sub_dst, sub_dst + sub_src], dtype=torch.long)
    labels_sample = labels_src[idx_sample]
    sub_data = Data(
        x=data_src.x[idx_sample],
        edge_index=sub_edge_index,
        y=torch.tensor(labels_sample, dtype=torch.long)
    )
    return sub_data, idx_sample, labels_sample


def draw_graph_frame(ax, G, pos, labels_sample, colors, class_labels, title, node_size, edge_alpha, progress=1.0):
    ax.clear()
    ax.set_facecolor('#F7FBFF')
    ax.figure.patch.set_facecolor('white')

    if G.number_of_edges() > 0:
        nx.draw_networkx_edges(
            G, pos, ax=ax,
            edge_color='#B7C1CC',
            alpha=max(0.05, float(edge_alpha) * progress),
            width=0.45 + 0.25 * progress
        )

    glow_size = max(90, node_size * 18) * (0.65 + 0.35 * progress)
    node_size_current = max(45, node_size * 11) * (0.55 + 0.45 * progress)

    for cls_val, color in enumerate(colors):
        nodes_for_class = np.where(labels_sample == cls_val)[0].tolist()
        if not nodes_for_class:
            continue
        nx.draw_networkx_nodes(
            G, pos, nodelist=nodes_for_class, ax=ax,
            node_color=color,
            node_size=glow_size,
            alpha=0.08 + 0.08 * progress,
            linewidths=0
        )
        nx.draw_networkx_nodes(
            G, pos, nodelist=nodes_for_class, ax=ax,
            node_color=color,
            node_size=node_size_current,
            alpha=0.9,
            edgecolors='white',
            linewidths=0.5 + 0.3 * progress,
            label=class_labels[cls_val]
        )

    ax.set_title(title, fontsize=15, pad=12, color='#1F2937')
    ax.text(
        0.02, 0.02,
        'Render con matplotlib: halos, capas y revelado progresivo',
        transform=ax.transAxes,
        fontsize=9,
        color='#4B5563',
        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='#D1D5DB', alpha=0.9)
    )
    ax.axis('off')
    ax.legend(loc='upper right', frameon=True, facecolor='white', edgecolor='#D1D5DB')


def plot_graph_overview(case_key, n_nodes, k_neighbors, node_size, edge_alpha, animate_graph=True):
    import time

    data_src, labels_src, colors, class_labels, title_base = get_graph_source(case_key, k_neighbors)
    sub_data, idx_sample, labels_sample = build_connected_subgraph(data_src, labels_src, n_nodes)

    G = to_networkx(sub_data, to_undirected=True)
    total_nodes = G.number_of_nodes()
    total_edges = G.number_of_edges()
    k_layout = 1.2 / np.sqrt(total_nodes) if total_nodes else 0.12
    pos = nx.spring_layout(G, seed=42, k=k_layout, iterations=50)

    title = f'{title_base} ({total_nodes} nodos, k={k_neighbors})'
    animate_this = bool(animate_graph) and total_nodes <= 1000
    stages = [0.38, 0.68, 1.0] if animate_this else [1.0]

    fig, ax = plt.subplots(figsize=(11.6, 8.8))
    for frame_idx, progress in enumerate(stages):
        draw_graph_frame(
            ax, G, pos, labels_sample, colors, class_labels,
            title, node_size, edge_alpha, progress
        )
        clear_output(wait=True)
        display(fig)
        if animate_this and frame_idx < len(stages) - 1:
            time.sleep(0.12)
    plt.close(fig)

    avg_degree = (2 * total_edges / total_nodes) if total_nodes else 0.0
    animation_msg = 'Animacion suave activada.' if animate_this else 'Vista estatica renderizada.'
    if animate_graph and total_nodes > 1000:
        animation_msg = 'Animacion omitida para mantener fluidez con mas de 1000 nodos.'
    update_graph_info([
        f'Estadisticas del subgrafo conectado (BFS): {total_nodes} nodos | {total_edges} aristas | grado promedio: {avg_degree:.1f}',
        f'{animation_msg} Ajusta los controles y vuelve a pulsar Visualizar Grafo para regenerar.'
    ])


def on_graph_regen(b):
    if panel_state['graph_busy']:
        return

    panel_state['graph_busy'] = True
    graph_regen_btn.disabled = True

    with graph_output:
        clear_output(wait=True)
        try:
            case = graph_case_select.value
            n_nodes = graph_nodes_slider.value
            k_val = graph_k_slider.value
            n_size = node_size_slider.value
            e_alpha = edge_opacity_slider.value
            animate_graph = graph_animate_toggle.value

            if k_val != 5:
                print(f'Reconstruyendo grafo con k={k_val}... puede tardar unos segundos.')

            plot_graph_overview(case, n_nodes, k_val, n_size, e_alpha, animate_graph=animate_graph)
        except Exception as exc:
            print(f'No se pudo generar el grafo: {exc}')
            update_graph_info(['No hay estadisticas disponibles porque la visualizacion fallo.'])
        finally:
            graph_regen_btn.disabled = False
            panel_state['graph_busy'] = False


graph_regen_btn.on_click(on_graph_regen, remove=True)
graph_regen_btn.on_click(on_graph_regen)

update_graph_info(['Selecciona parametros, decide si quieres animacion y haz clic en Visualizar Grafo.'])

graph_tab_content = widgets.VBox([
    widgets.HBox([graph_case_select, graph_nodes_slider]),
    widgets.HBox([graph_k_slider, node_size_slider]),
    widgets.HBox([edge_opacity_slider, graph_animate_toggle, graph_regen_btn]),
    graph_info_output,
    graph_output
], layout=widgets.Layout(padding='10px'))

with graph_output:
    try:
        plot_graph_overview('c1', 800, 5, 8, 0.15, animate_graph=True)
    except Exception as exc:
        print(f'No se pudo cargar el grafo inicial: {exc}')
        update_graph_info(['No hay estadisticas disponibles porque la visualizacion inicial fallo.'])

tabs = widgets.Tab()
tabs.children = [pred_tab_content, graph_tab_content]
tabs.set_title(0, 'Prediccion en Vivo')
tabs.set_title(1, 'Explorador de Grafos')

panel_title = widgets.Label(value='SentinelGrafo - Panel Interactivo')
display(panel_title)
display(tabs)
```
